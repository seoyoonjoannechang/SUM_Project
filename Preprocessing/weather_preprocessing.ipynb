# 피처 설명

- tm (일시) 
    - 관측 날짜와 시간
    
    
- ta (기온, °C)


- rn (강수량, mm)
    - 비가 오지 않거나 아주 조금 온 경우 = 0
    
    
- ws (풍속, m/s)


- wd (풍향, 16방위)
<img src = "./data/project_data/wd.jpg" height = 1000 weight = 1200>

- hm (습도, %)
    - 관측 시 대기 중에 실제로 포함되어 있는 수증기의 양(수증기압)과 그때의 기온으로 포함시킬 수 있는 최대 수증기량(포화수증기압)과의 비를 상대습도라고 하며 백분율(%)로 표현
    
    
- pv (증기압, hPa)
    - 증기가 고체 또는 액체와 동적 평형 상태에 있을 때 증기의 압력
    
    
- td (이슬점온도, °C)
    - 공기가 포화되어 수증기가 응결할 때의 온도
    
    
- pa (현지기압, hPa)
    - 관측소 높이에서의 기압


- ps (해면기압, hPa)
    - 어떤 관측지점에서 관측된 관측소 기압에 해면경정을 하여 얻은 평균해수면 높이에서의 기압
    
    
- ss (일조, hr)
    - 태양의 직사광이 지표면에 비친 시간
    
    
- icsr (일사, MJ/m2)
    - 태양의 복사에너지가 땅에 닿았을 때의 세기
    
    
- dsnw (적설, cm)
    - 눈, 싸락눈 등이 지면에 내려 쌓여 있는 수직 깊이
    - 눈이 관측장소 또는 관측장소 주위의 지면에 반 이상을 덮었을 때를적설이 있는 것
    - 많은 눈이 내리더라도 지면에 쌓이기 전에 녹아지면(관측장소 또는 청사주위)의 반 이상을 덮지 못하면 적설은 없는 것으로 간주
    
- hr3Fhsc (3시간신적설, cm)
    
    
- dc10Tca (전운량, 10분위)
    - 천공에 전혀 구름이 없거나 있더라도 5% 미만일 때의 운량은 0이다.
    - 천공이 구름으로 전부 덮여 있거나 95 % 이상 덮여 있을 때의 운량은 10이다. 
    - 운량은 0에서 l0사이의 정수로 나타낸다. 
    
    
- dc10LmcsCa (중하층운량, 10분위)
    - 전 하늘을 10으로 하여 구름이 덮은 부분을 10분위의 2자리로 관측한다.
    
    
- clfmAbbrCd (운형, 운형약어)
    - 10가지 기본 운형을 높이에 따라 상,중,하로 구분하여 관측
    
    
- lcsCh (최저운고, 100m)
    - 운고는 관측장소의 지면에서부터 구름 밑면까지의 높이
    - 구름의 높이를 100m 단위 두 자리 수로 관측하고 기록한다.
    
    
- vs (시정, 10m)
    - 대기의 혼탁 정도를 나타내는 기상요소, 정상적인 시각을 가진 사람이 식별할 수 있는 최대 거리
    
    
- gndSttCd (지면상태코드)

### 결측치 대체

- rn, dsnw
    - 비와 눈이 안 올 경우 Nan
    - 0으로 처리


- ss, icsr
    - 태양에 영향이 있으므로 태양이 없는 시간대에는 Nan
    - 0으로 처리


- dmstMtphNo 
    - 현상번호로 여러 기상현상을 나타냄
    - 기상현상이 안 일어난 경우 Nan
    - 0으로 처리


- lcsCh 
    - 구름의 영향이 없는 경우 Nan
    - 0으로 처리
    
    
- 나머지 피처는 연속적인 피처로 판단, 앞 뒤 값의 평균으로 대체

```python
# 결측치 처리

def weather_preprocessing(df):
    
    # 1. 결측치 대체
    cols_for_zero = ["rn","dsnw","ss","icsr","dmstMtphNo","lcsCh"]
    cols_for_inter = ['ta','ws','wd','hm','pv','td','pa','ps','dc10Tca','dc10LmcsCa','vs','ts']
                      
    df = globals()["{}".format(df)]
    columns = df.columns
    print(columns)
    flg_for_drop = []
    for column in columns:
        if "Qcflg" in column:
            flg_for_drop.append(column)
    cols_for_drop = ["rnum", "stnId","hr3Fhsc", "clfmAbbrCd", "gndSttCd", "m005Te", "m01Te", "m02Te", "m03Te"]
    
    # 결측치 0으로 처리
    for column in cols_for_zero:
        df[column].fillna(0, inplace = True)
    
    # 결측치 앞 뒤 평균값 처리
    for column in cols_for_inter:
        df[column].interpolate(inplace = True)
    
    # 결측치 제거
    for column in cols_for_drop :
        df.drop(column, axis = 1, inplace = True)
    
    for column in flg_for_drop :
        df.drop(column, axis = 1, inplace = True)
        
    #2. "tm" 열이름 "datetime" 으로 변경
    df.rename(columns = {"tm" : "datetime"}, inplace = True)
    
    return df
```

# 데이터 datetime 수 맞추기

- 파주는 77개의 datetime이 누락되있음
- 없는 날짜 데이터는 바로 뒤 데이터로 채운다

```python
def df_shape(var) :    
    var["datetime"] = pd.to_datetime(var["datetime"])
    ndf = var.set_index("datetime")
    ndf = ndf.asfreq(freq = "1h")
    ndf = ndf.reset_index()
    ndf = ndf.fillna(method = "bfill")
    ndf["datetime"] = ndf["datetime"].astype(str)
    return ndf
```
